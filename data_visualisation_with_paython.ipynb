Section 2- DATA ANALYSIS AND VISUALIZAION


```python
#IMPORT LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Creating a Dataset
```

```python

salesdata={'Date': pd.date_range(start='2023-01-01', periods=100, freq='D'), 'Stores': ['St A', 'St B', 'St C', 'St D']* 25, 'Product':['Prod 1', 'Prod 2', 'Prod 3', 'Prod 4']*25, 'Sales': np.random.randint(50,200, size=100), 'Revenue':  np.random.randint(1000,50000, size=100) }
df=pd.DataFrame(salesdata)
df
```

```python
#Data Transformation
df['Date']=pd.to_datetime(df['Date'])
```

```python

# Sales over the time plot
plt.figure(figsize=(14,8))
sns.lineplot(data=df, x='Date', y='Sales')
plt.title('Sales over time')
plt.xlabel('Dates')
plt.ylabel('Sales')
```

```python
# bar plot as sales across stores
plt.figure(figsize=(14,8))
sns.barplot(data=df, x='Stores', y='Sales', estimator=sum)
plt.title('Sales over time')
plt.xlabel('Stores')
plt.ylabel('Sales')
```

```python
sales_trend=df.groupby (df['Date'].dt.month) ['Sales'].sum()
sales_trend.plot(kind='bar', figsize=(12,6))
plt.title('Monthly sales')
plt.xlabel('Month')
plt.ylabel('Total sales')
```
